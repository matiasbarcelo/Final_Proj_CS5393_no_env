```python
# import spacy
# from spacy.tokens import DocBin
# from spacy.util import minibatch, compounding

# def train_ner_model(train_data, iterations=30):
#     nlp = spacy.blank("en")
#     ner = nlp.add_pipe("ner")
#     for _, annotations in train_data:
#         for ent in annotations.get("entities"):
#             ner.add_label(ent[2])

#     other_pipes = [pipe for pipe in nlp.pipe_names if pipe != "ner"]
#     with nlp.disable_pipes(*other_pipes):
#         optimizer = nlp.begin_training()
#         for itn in range(iterations):
#             random.shuffle(train_data)
#             losses = {}
#             batches = minibatch(train_data, size=compounding(4., 32., 1.001))
#             for batch in batches:
#                 texts, annotations = zip(*batch)
#                 nlp.update(texts, annotations, drop=0.5, losses=losses)
#             print(f"Iteration {itn}, Losses: {losses}")

#     return nlp

# # Train the model with your domain-specific data

import spacy
from spacy.tokens import DocBin
from spacy.training.example import Example
from spacy.util import minibatch, compounding
import random

def train_ner_model(train_data, iterations=30):
    # Create a blank English model
    nlp = spacy.blank("en")
    # Add the NER pipeline
    ner = nlp.add_pipe("ner")
    # Add entity labels to the NER model
    for _, annotations in train_data:
        for ent in annotations.get("entities"):
            ner.add_label(ent[2])

    # Disable other pipeline components during training
    other_pipes = [pipe for pipe in nlp.pipe_names if pipe != "ner"]
    with nlp.disable_pipes(*other_pipes):
        optimizer = nlp.begin_training()
        for itn in range(iterations):
            random.shuffle(train_data)
            losses = {}
            batches = minibatch(train_data, size=compounding(4.0, 32.0, 1.001))
            for batch in batches:
                examples = []
                for text, annotations in batch:
                    # Convert text and annotations into Example objects
                    doc = nlp.make_doc(text)
                    example = Example.from_dict(doc, annotations)
                    examples.append(example)
                # Update the model
                nlp.update(examples, drop=0.5, losses=losses)
            print(f"Iteration {itn}, Losses: {losses}")

    return nlp

```

```python
import os
import re

def clean_text_file(input_path, output_path):
    """
    Cleans a .txt file extracted from a PDF.
    Args:
        input_path (str): Path to the input .txt file.
        output_path (str): Path to save the cleaned .txt file.
    """
    try:
        with open(input_path, "r", encoding="utf-8") as file:
            text = file.read()
        
        # Remove multiple spaces and newlines
        text = re.sub(r'\s+', ' ', text)

        # Remove non-UTF characters (optional)
        text = re.sub(r'[^\x00-\x7F]+', '', text)

        # Remove page numbers (e.g., patterns like "Page 1" or "1")
        text = re.sub(r'(Page\s?\d+|\b\d+\b)', '', text)

        # Normalize spacing around punctuation
        text = re.sub(r'\s+([.,!?;])', r'\1', text)

        # Convert to lowercase (optional)
        # text = text.lower()

        # Trim leading and trailing spaces
        text = text.strip()

        # Save cleaned text
        with open(output_path, "w", encoding="utf-8") as file:
            file.write(text)

        print(f"Cleaned text saved to {output_path}")
    except Exception as e:
        print(f"Error cleaning file {input_path}: {e}")


# Process all .txt files in a folder
def clean_all_text_files(folder_path):
    """
    Cleans all .txt files in a given folder.
    Args:
        folder_path (str): Path to the folder containing .txt files.
    """
    for filename in os.listdir(folder_path):
        if filename.endswith(".txt"):
            input_path = os.path.join(folder_path, filename)
            output_path = os.path.join(folder_path, f"cleaned_{filename}")
            clean_text_file(input_path, output_path)

# Example usage:
# folder_path = "/Users/anirudhvalluru/Desktop/pypractice/AdvPython/asoiaf copy"  # Folder with your .txt files
# clean_all_text_files(folder_path)
```

```python
import spacy
import networkx as nx

# Load your model
nlp = spacy.load("./domain_ner_model")

# Add the sentencizer to the pipeline to handle sentence segmentation
if 'sentencizer' not in nlp.pipe_names:
    # Directly add the sentencizer component to the pipeline
    nlp.add_pipe("sentencizer")

def extract_relationships(doc):
    relationships = []
    for sent in doc.sents:
        root = sent.root
        subject = None
        obj = None
        for child in root.children:
            if child.dep_ == "nsubj":
                subject = child
            if child.dep_ in ["dobj", "pobj"]:
                obj = child
        if subject and obj:
            relationships.append((subject, root, obj))
    return relationships

def process_document(text):
    doc = nlp(text)
    entities = [(ent.text, ent.label_) for ent in doc.ents]
    relationships = extract_relationships(doc)
    return entities, relationships

def build_knowledge_graph(documents):
    G = nx.DiGraph()
    for doc in documents:
        entities, relationships = process_document(doc)
        for entity, entity_type in entities:
            G.add_node(entity, type=entity_type)
        for subj, pred, obj in relationships:
            G.add_edge(subj.text, obj.text, relation=pred.text)
    return G

# Split a large text into smaller chunks (by paragraphs, sentences, etc.)
def chunk_text(text, chunk_size=1000):  # Adjust chunk_size based on your needs
    chunks = []
    for i in range(0, len(text), chunk_size):
        chunks.append(text[i:i+chunk_size])
    return chunks

# Load the book as a string (assuming 'book.txt' is the full book)
with open("/Users/anirudhvalluru/Desktop/pypractice/AdvPython/asoiaf copy/cleaned_GOT.txt", "r", encoding="utf-8") as file:
    full_text = file.read()

# Split the book into chunks
chunks = chunk_text(full_text)

# Process the chunks and build the knowledge graph
knowledge_graph = build_knowledge_graph(chunks)

# Visualize the graph or use it for further analysis
import matplotlib.pyplot as plt
nx.draw(knowledge_graph, with_labels=True, node_color="skyblue", edge_color="gray")
plt.show()
```
